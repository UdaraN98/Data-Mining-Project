```python
import os
import numpy as np

def extract_npy_files(dir1, dir2):

    data = []
    
    for dir_path, label in [(dir1, 0), (dir2, 1)]:
        for folder_name in os.listdir(dir_path):
            folder_path = os.path.join(dir_path, folder_name)
            if os.path.isdir(folder_path):
                fnc_path = os.path.join(folder_path, "fnc.npy")
                icn_tc_path = os.path.join(folder_path, "icn_tc.npy")
                if os.path.exists(fnc_path) and os.path.exists(icn_tc_path):
                    fnc = np.load(fnc_path)
                    icn_tc = np.load(icn_tc_path)
                    data.append((fnc, icn_tc, label))
    
    return data
```

```python
dir1 = "/content/kaggle_data/data/train/BP"
dir2 = "/content/kaggle_data/data/train/SZ"
data = extract_npy_files(dir1, dir2)
#data[0][0] shape = 5640*1
#data[0][1] shape = (230*)*105
#data[0][2] shape = label
```

```python
for i in range(25):
  print(data[i][2])
```

```python
import numpy as np
from keras.models import Model
from keras.layers import Input, Dense, Concatenate





# Shuffle the data
#np.random.shuffle(data)

# Split the data into train and validation sets
split = int(len(data) * 0.8)
train_data = data[:split]
val_data = data[split:]

# Define the inputs
fnc_input = Input(shape=(train_data[0][0].shape[1],))
icn_tc_input = Input(shape=(train_data[0][1].shape[1],))

# Define the shared layers
shared_layer1 = Dense(64, activation="relu")
shared_layer2 = Dense(32, activation="relu")

# Define the individual branches
fnc_branch = shared_layer1(fnc_input)
fnc_branch = shared_layer2(fnc_branch)

icn_tc_branch = shared_layer1(icn_tc_input)
icn_tc_branch = shared_layer2(icn_tc_branch)

# Concatenate the branches
concatenated = Concatenate()([fnc_branch, icn_tc_branch])

# Define the output layer
output_layer = Dense(1, activation="sigmoid")(concatenated)

# Define the model
model = Model(inputs=[fnc_input, icn_tc_input], outputs=output_layer)

# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Train the model
train_X_fnc = np.array([d[0] for d in train_data])
train_X_icn_tc = np.array([d[1] for d in train_data])
train_y = np.array([d[2] for d in train_data])
val_X_fnc = np.array([d[0] for d in val_data])
val_X_icn_tc = np.array([d[1] for d in val_data])
val_y = np.array([d[2] for d in val_data])
model.fit([train_X_fnc, train_X_icn_tc], train_y, epochs=10, batch_size=32, validation_data=([val_X_fnc, val_X_icn_tc], val_y))
```
