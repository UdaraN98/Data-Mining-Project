Upsampling the dataset to improve the accuracy of the previous RF model


```python
#importing the pre processed datasets as pandas dataframes

import pandas as pd

train = pd.read_csv("TrainV1.csv")
test = pd.read_csv("TestV1.csv")
```

```python
#counting the number of elements in each class in the target classes

train["attempts_range"].value_counts()
```

```python
#partitioning the training data
y=train.attempts_range
x=train.drop('attempts_range',axis=1)
```

```python

#label encoding the y
from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
```

```python
#using SMOTE to oversampling

from imblearn.over_sampling import SMOTE

OS = SMOTE()
x,y = OS.fit_resample(x,y)
```

```python
#checking the number of items in each class
from collections import Counter

cnt = Counter(y)
print(cnt)
```

```python
#As the dataset is balanced, model inititiation and training

from sklearn.ensemble import RandomForestClassifier

model0 = RandomForestClassifier(max_depth = 24, random_state = 0)
model0.fit(x, y)
```

```python
#predintion on train dataset
pred0  = model0.predict(x)
#checking the F1 score of the model

from sklearn.metrics import f1_score, accuracy_score

f1 =f1_score(y,pred0, average = "weighted")

acc =accuracy_score(y, pred0)

print("Accuracy:" ,acc)
print("F1 :",f1)
```

Creating a Gradient Boosted Tree algorithm as a candidate model
