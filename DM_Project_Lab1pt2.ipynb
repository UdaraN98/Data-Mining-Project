### **Merging the three datasets and exploring for null values and zero values**

```python
#importing the relevant modules

import pandas as pd
import numpy as np
```

```python
#opening the csv files as pandas dataframes

problems_DF = pd.read_csv("problem_data.csv")
train_DF = pd.read_csv("train_submissions.csv")
user_DF =  pd.read_csv("user_data.csv")
```

```python
#creating a one dataframe by merging the datasets

Df1 = pd.merge(problems_DF,train_DF,on="problem_id")
Df2 = pd.merge(Df1,user_DF, on= "user_id")
```
